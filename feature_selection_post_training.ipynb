### Scale the data

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler()

scaler = MinMaxScaler()
feature_columns = df.columns[1:24]
label = df.columns[24:25]
#print(feature_columns)
new_test_features = test[feature_columns]
new_test_labels = test[label]
new_test_features[feature_columns] = scaler.fit_transform(new_test_features[feature_columns])
df[feature_columns] = scaler.fit_transform(df[feature_columns])
df.head()
```

### Split the data

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

# First extract our test data and store it in x_test, y_test
features = df[feature_columns].to_numpy()
labels = df[label].to_numpy()
_x, x_test, _y, y_test = train_test_split(features, labels, test_size=0.10, random_state=42) #_x is the rest of the data

# set k = 5
k = 5

kfold_spliiter = KFold(n_splits=k)

folds_data = [] # this is an inefficient way but still do it

fold = 1
for train_index, validation_index in kfold_spliiter.split(_x):
    x_train , x_valid = _x[train_index,:],_x[validation_index,:]
    y_train , y_valid = _y[train_index,:] , _y[validation_index,:]
    print (f"Fold {fold} training data shape = {(x_train.shape,y_train.shape)}")
    print (f"Fold {fold} validation data shape = {(x_valid.shape,y_valid.shape)}")
    fold+=1
    folds_data.append((x_train,y_train,x_valid,y_valid))
```

### Train the model

```python
# Now let's define our models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier


# LR with no regularizer
lr_vanilla = LogisticRegression(penalty='none')
lr_L2 = LogisticRegression(penalty="l2")
svm_linear = SVC(kernel="linear")
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(random_state=23) # some random seed for reproducibility
grad_boost = GradientBoostingClassifier()
voting = VotingClassifier(estimators=[("1",lr_vanilla),("2",svm_linear),("3",dt)])



svm_linear = SVC(kernel="linear")
svm_poly = SVC(kernel="poly",degree=2)

# Keep all the models in a dictionary

all_models = {"lr_vanilla":lr_vanilla, 
              "lr_L2":lr_L2,
              "svm_linear":svm_linear,
              "svm_poly":svm_poly,
             "decision_tree":dt,
              "random_forest":rf,
              "grad_boost":grad_boost,
              "voting":voting}
```

```python
best_validation_accuracy = 0
best_model_name = ""
best_model = None

# Iterate over all models
for model_name in all_models.keys():
    
    print (f"Evaluating {model_name} ...")
    model = all_models[model_name]
    
    # Let's store training and validation accuracies for all folds
    train_acc_for_all_folds = []
    valid_acc_for_all_folds = []
    
    #Iterate over all folds
    for i, fold in enumerate(folds_data):
        x_train, y_train, x_valid, y_valid = fold

        # Train the model
        _ = model.fit(x_train,y_train.flatten())

        # Evluate model on training data
        y_pred_train = model.predict(x_train)
        
        # Evaluate the model on validation data
        y_pred_valid = model.predict(x_valid)
        
        # Compute training accuracy
        train_acc = accuracy_score(y_pred_train , y_train.flatten())
        
        # Store training accuracy for each folds
        train_acc_for_all_folds.append(train_acc)
        
        # Compute validation accuracy
        valid_acc = accuracy_score(y_pred_valid , y_valid.flatten())

        # Store validation accuracy for each folds
        valid_acc_for_all_folds.append(valid_acc)
    
    #average training accuracy across k folds
    avg_training_acc = sum(train_acc_for_all_folds)/k
    
    print (f"Average training accuracy for model {model_name} = {avg_training_acc}")
    
    #average validation accuracy across k folds
    avg_validation_acc = sum(valid_acc_for_all_folds)/k
    
    print (f"Average validation accuracy for model {model_name} = {avg_validation_acc}")
    
    # Select best model based on average validation accuracy
    if avg_validation_acc > best_validation_accuracy:
        best_validation_accuracy = avg_validation_acc
        best_model_name = model_name
        best_model = model
    print (f"-----------------------------------")

print (f"Best model for the task is {best_model_name} which offers the validation accuracy of {best_validation_accuracy}")
```

## Feature Selection 
We were able to rank features based on ablation studies. Let's systematically go through other methods that are used for selecting the best features. 

### Model and label agnostic feature selection
Methods under this category only work with feature columns and typically used to throw away random, noisy and non-informative features. This could be very useful especially in **unsupervised learning**, because we don't have labels there. 

#### Correlation based filtering
The idea here is that if two features are extremely correlated, we can get rid of one of them, thereby reducing feature redundancy.

### 2.1.2. Variance thresholding
The idea is that if a feature exhibits low variance, it is not very informative. Why?

Consider this example. You are trying to predict whether a person will be diabetic or not and you have a column "human or not". This will obviously have no variance (since all our datapoints are for humans). 

In our Mt Rainer exercise, we had `battery_avg` which exhibited similary properties. All dry cell batteries have the save voltage label (and thereby, very low variance). We can get rid of such features. 

In geenral, we can consider features if they exhibit normalized variance greater than a certain threshold. 

```python
from sklearn.feature_selection import VarianceThreshold

test_features = x_test
test_features = pd.DataFrame(np.array([x for x in test_features]))
selector = VarianceThreshold()
train_features.columns = list(df.columns[1:24])
test_features.columns = list(df.columns[1:24])
selector.fit(train_features)

print(train_features)

print(selector.feature_names_in_)
print(selector.variances_)

selector = VarianceThreshold(threshold=0.02)
selector.fit(train_features)
print (f"Selected features: {selector.get_feature_names_out()}")

train_features_reduced = selector.transform(train_features)
test_features_reduced = selector.transform(new_test_features)
```

## 2.2. Model agnostic but label aware feature selection 
We aim to select the features which are highly dependent on the label.

### 2.2.1. Chi-squared test 
Chi-squared test is a commonly used mechanism to accept or reject that two variable distributions are equal or not. If they are very different, the chi-squared score will be significantly different from the ideal value, suggesting that the features and labels are completely independent. In this case, we can remove the feature. 

Refer to: https://en.wikipedia.org/wiki/Chi-squared_test


```python
from sklearn.feature_selection import chi2

chi_scores = chi2(train_features,_y)
print(chi_scores)
print ("Ranking based on chi-squared test")

score_list = []
for i, feature in enumerate(feature_columns):
    score_list.append([chi_scores[1][i],feature])

# print(score_list)
sorted_score_list = sorted(score_list)

print(f"Printing ranked features based on chi-squared value")
print("lower p-value indicates feature and label are dependent)")
# print(sorted_score_list)
for score, feature in sorted_score_list:
    print (f"Feature {feature} : p-value {score}")
    
```

## 2.3. Model Aware Feature Seleciton

This way of feature selection assumes that we will select features based on an underlying ML model. Feature ablation study is an example of model aware feature selection. 

We can select feature by applying L1 regularization on a parametric model (such as SVM or LR). This is called the `Lasso` method. 

## 2.3.1 Lasso Regularization on the linearSVC then apply feature selection result to the random forest model

Since lasso minimizes a sum of absolute values of the weights to the objective function, it ends up either retaining or getting rid of the features, thus helping in feature selection

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.svm import LinearSVC

base_model = LinearSVC(C=0.1, penalty="l1", dual=False)
base_model.fit(train_features.to_numpy(),_y)
selector = SelectFromModel(base_model, prefit=True)

train_features_reduced = selector.transform(train_features.to_numpy())
test_features_reduced = selector.transform(new_test_features.to_numpy())

# print(len(train_features_reduced))
selected_feature_indices = selector.get_support(indices=True)
selected_feature_names = train_features.columns[selected_feature_indices]
# print(selected_feature_indices)
print(selected_feature_names)
print(len(train_features_reduced))


rf.fit(train_features_reduced ,_y)
# now predict and compute acuracy
test_actual = new_test_labels
# print(len(test_actual))
# print(test_features_reduced)
test_predicted = rf.predict(test_features_reduced)

print (classification_report(test_actual,test_predicted))
```
